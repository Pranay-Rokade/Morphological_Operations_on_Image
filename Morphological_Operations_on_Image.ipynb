```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = 'leg.jpg'
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img, cmap='gray')
plt.axis("off")
plt.show()
```

```python
# Apply binary thresholding
_, binary_image = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV) # white pixels have
```

```python
plt.imshow(binary_image, cmap='grey')
plt.axis("off")
plt.show()
```

```python
# Define a kernel for the morphological operations
kernel = np.ones((5, 5), np.uint8) # kernel = mask(3*3)=filter=moving across the img for the operation
```

```python
# 1. Dilation
dilated_image = cv2.dilate(binary_image, kernel, iterations=1)
# The kernel slides over the binary_image and for each position,
# if any pixel within the kernel is white (255), the central pixel (where the kernel is positional) will be set to white (255)
plt.imshow(dilated_image, cmap='gray')
plt.axis("off")
plt.show()
```

```python
# 2. Erosion
eroded_image = cv2.erode(binary_image, kernel, iterations=1)
plt.imshow(eroded_image, cmap='gray')
plt.axis("off")
plt.show()
```

```python
# 3. Opening
opened_image = cv2.morphologyEx(binary_image, cv2.MORPH_OPEN, kernel)
plt.imshow(opened_image, cmap='gray')
plt.axis("off")
plt.show()
```

```python
# 4. Closing
closed_image = cv2.morphologyEx(binary_image, cv2.MORPH_CLOSE, kernel)
plt.imshow(closed_image, cmap='gray')
plt.axis("off")

plt.show()
```

```python
# 5. Gradient
gradient_image = cv2.morphologyEx(binary_image, cv2.MORPH_GRADIENT, kernel)
plt.imshow(gradient_image, cmap='gray')
plt.axis("off")
plt.show()
```

```python
# Display the results using matplotlib

plt.figure(figsize=(10, 10))
# Display each image
titles = ['Original', 'Binary', 'Dilated', 'Eroded', 'Opened', 'Closed', 'Gradient']
images = [img, binary_image, dilated_image, eroded_image, opened_image, closed_image, gradient_image]

for i in range(len(images)):
    plt.subplot(3, 4, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off') 
plt.tight_layout()
plt.show()
```
